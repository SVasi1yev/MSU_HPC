```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
a1 = [1, 0.9779, 0.6492, 0.4821]
a2 = [1, 0.9834, 0.5923, 0.4717]
a3 = [1, 1.0162, 0.8286, 0.8844]
```

```python
font = {'size': 20}
plt.rc('font', **font)
fig = plt.figure(figsize=(14, 10))
plt.grid()
plt.xticks([0, 2, 4, 16, 32])

p = [2, 4, 16, 32]
plt.plot(p, a1, marker='o', label='err = 3.0 * 1e-5')
plt.plot(p, a2, marker='o', label='err = 5.0 * 1e-6')
plt.plot(p, a3, marker='o', label='err = 1.5 * 1e-6')
plt.legend()
plt.xlabel('Число процессов')
plt.ylabel('Ускорение')
```
